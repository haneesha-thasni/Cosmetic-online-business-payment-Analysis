# Cosmetic Online Business Payment Data (2021)
This dataset contains real-world payment transactions from an online cosmetic business that operated in 2021. The data captures various aspects of the transactions, including payment amounts, types, providers, and geographic distribution. Additionally, it includes details about the orders, such as the products purchased, quantities, discounts applied, and shipping costs.

The dataset provides a rich source of information for analyzing consumer behavior, payment trends, and order dynamics in the e-commerce sector, particularly within the cosmetics industry.

# Key Features:

* Time & Date: Timestamp of the payment.
* Currency: Currency in which the payment was made (e.g., INR).
* Amount: The amount paid for the order.
* Payment Type: The type of payment method used.
* Payment Provider: The service provider facilitating the payment (e.g., Razorpay, Offline Payments).
* City: The city from which the payment was made.
* Order Type: The platform or medium used to place the order.
* Order ID: A unique identifier for each order.
* Name: Description of the product(s) in the order.
* Quantity: Number of units purchased.
* Discount: Discount applied to the order, if any.
* Shipping: Shipping cost associated with the order.
* Status: Status of the payment (e.g., Successful).

# Annual Comparison of Payment Amounts Processed by Payment Provider

```python
sns.barplot(x='Year',y='Amount',data=df,hue='Payment Provider')
plt.savefig('payment provider.png',dpi=300)
plt.show()
```

* ###### Up until 2021, offline payments were the most commonly used payment method, with Razorpay adoption being relatively low in comparison. However, starting in 2021 and continuing into 2022, there was a significant shift, with Razorpay being increasingly preferred over offline payments. By 2023, Razorpay had become the only payment method used, with no recorded instances of offline payments.

#### Possible Reasons:

* Digital Transformation and Pandemic Impact: 

  The COVID-19 pandemic accelerated the shift to digital payment methods as consumers and businesses increasingly moved online, reducing reliance on cash and offline transactions.
* Convenience of Online Payments: 

  Platforms like Razorpay offer seamless, secure, and faster transactions, making them more attractive than traditional offline payments, especially for e-commerce businesses.
* Increased Trust in Digital Payment Security: 

  As digital payment systems became more secure and reliable, more businesses and customers adopted online payment methods over traditional ones.
* Government Push for Digitalization: 

  The Indian government’s push for a cashless economy and initiatives like Digital India further encouraged the adoption of online payments like Razorpay.
* Merchant and Consumer Preferences: 

   As more merchants began offering Razorpay as a payment option, consumers grew more accustomed to and preferred the convenience of using Razorpay over offline methods.

# Overall Findings

* #### Currency Usage: 
All transactions in the dataset were conducted in Indian Rupees (INR), with no indication of multi-currency usage, showing a localized business operation within India.

* #### Cosmetic Payments Surge in 2021: 
There was a sharp increase in cosmetic online payments in 2021, likely driven by the COVID-19 pandemic, which encouraged people to shop online. More time at home and personal care needs may have led to increased interest in cosmetics, bolstered by promotions and discounts.

* #### Decline in 2022 Payments: 
After the 2021 peak, there was a decline in cosmetic online payments in 2022, possibly due to market saturation or changes in consumer habits as people returned to pre-pandemic lifestyles. Another factor could be the growing preference for alternative purchasing channels or products.

* #### Monthly Payment Trends: 
May and April were the most active months for cosmetic payments, probably due to seasonal promotions and increased personal care needs during the summer season. February saw the fewest payments, indicating lower activity at the beginning of the year.

* #### Day-wise Payment Trends: 
Wednesday saw the highest number of payments, suggesting that mid-week transactions are popular, possibly because businesses tend to settle accounts. In contrast, Friday had the lowest payment count, likely due to end-of-week delays or preference for processing payments the following week.

* #### City-wise Payment Insights: 
Delhi and Mumbai were the top markets for cosmetic online businesses, with 112 and 106 payments, respectively, highlighting their dominant positions. Other significant cities included Nagpur, Kolkata, Bengaluru, and Hyderabad.

* #### Order Quantity by City: 
In terms of order quantity, Agartala led with 33 orders, followed by Ahmedabad with 24 orders. Smaller cities like Adra and Ahmednagar had fewer orders, demonstrating uneven distribution of orders across cities.

* #### Payment Success Rate: 
The overall payment system was efficient, with a 96.89% success rate for transactions, 3.02% declined payments, and 0.09% refunded transactions, indicating minimal issues in payment processing.

* #### Payment Method Shift (Offline to Online): 
Before 2021, offline payments dominated, but the landscape changed with a notable shift towards Razorpay in 2021 and 2022. By 2023, Razorpay became the only payment method used, largely due to factors like digital transformation, pandemic-related shifts to e-commerce, government support for cashless transactions, and increased consumer trust in online payments.

* #### Shipping and Discount Trends: 
The relationship between the quantity of items shipped and shipping cost showed no strong correlation, indicating that shipping cost was likely determined by factors other than the quantity. Larger discounts were generally offered for higher quantities, but the relationship wasn’t consistent, suggesting that discount strategies varied depending on other factors such as promotions, product type, or customer segment.

* #### Payment Distribution: 
The data revealed that most payments were of smaller amounts, with fewer high-value transactions, as indicated by the right-skewed distribution. This suggests that cosmetic purchases tend to involve lower-cost items with occasional larger transactions.



# Recommendations

* #### Focus on Key Markets: 
With Delhi and Mumbai being the largest markets for cosmetic payments, businesses should focus on enhancing their marketing efforts in these regions. Cities like Nagpur and Kolkata also show potential for growth and should be prioritized for targeted promotions.

* #### Revitalize Interest Post-2022: 
Given the decline in payments after 2021, businesses should focus on strategies to reinvigorate consumer interest, such as launching new products, offering exclusive discounts, or bundling products for better value. Understanding shifts in consumer behavior post-pandemic can help refine these approaches.

* #### Capitalize on High-Activity Months: 
Leverage the surge in payments during April and May by planning seasonal campaigns or holiday promotions. These efforts can drive consumer engagement during peak periods, ensuring maximum sales impact.

* #### Mid-Week Promotions: 
Since Wednesday experiences the highest volume of payments, businesses should plan mid-week promotions or offer special deals that encourage even more transactions on this day. End-of-week promotions could also help boost Friday payments.

* #### Adopt Digital Payment Methods: 
Given the complete transition to Razorpay by 2023, businesses should continue to adopt digital payment platforms and ensure seamless, secure online transactions. This will further build consumer trust and facilitate smoother transactions.

* #### Improve Small-Ticket Item Strategies: 
The right-skewed distribution suggests that the majority of purchases are low-value. To capitalize on this, businesses can offer upselling options for larger bundles or product combinations that provide added value to customers, encouraging larger purchases.

* #### Fine-tune Shipping and Discount Strategies: 
Review and optimize the shipping cost model to ensure it's competitive and encourages more orders. Also, create more consistent discount strategies for high-quantity orders to incentivize larger purchases and repeat business.

* #### Monitor Payment Success Rates: 
While the success rate of payments is high, continued efforts should be made to minimize declined payments by improving the payment experience (e.g., addressing technical issues) and offering more flexible payment options to reduce processing failures.
